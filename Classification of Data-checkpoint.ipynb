# Rejection sampling algorithm#

Given expected average acceptance probability which is equal to $$\frac{1}{M} \overset{M=1.5}{=} \frac{1}{1.5} = 0.(6)$$.

Rejection sampling algorithm can be implemented below  sampling from $\mu$ using the
reference measure n = Unif[0, 1]. to solve the following tasks: 
Draw 10'000 samples from $\mu$ and plot a histogram of them together with a plot
of the density $\rho_\mu$. The shapes should match closely enough.

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10000 #  will generate 10'000 samples from a distribution

x = np.linspace(0,1,100) # taking the grid points x in order to plot functions rho_mu and rho_nu
rho_mu = 6*x*(1-x) # values of Cx(1-x) where C=6 (such that the area under the parabola on [0,1] is 1)
rho_nu = np.ones(x.shape) # uniform distribution in [0,1], which is equal to 1 everywhere

M = 1.5 # the least constant that limits r(x) = ro_mu / ro_nu for any x

samples_array = []
n_accepted = 0 # amount of accepted samples
n_rejected = 0 # amount of rejected samples
n_total = 0 # total amount of samples drawn

for i in range(N):
    while True:
        x_sample = np.random.uniform(0,1) # take a random number from uniform distribution nu (which is [0,1])
        u = np.random.uniform(0,1) # take a random number from uniform distribution [0,1]
        n_total+=1
        r = 6*x_sample*(1-x_sample) #calculate r as ro_mu / ro_nu, but since ro_nu==1 everywhere, 
                                    # it is the same as dividing by 1
        alpha = r/M # calculate alpha(x)
        if u <= alpha:
            n_accepted+=1
            samples_array.append(x_sample)
            break
        else:
            n_rejected +=1

samples_array = np.array(samples_array)

plt.figure(figsize=(10,8))
plt.plot(x, rho_mu, label ='rho_mu')
plt.plot(x, rho_nu,label ='rho_nu (uniform)')
plt.hist(samples_array, bins = 20, density=True, label = 'histogram of samples (normalized, \n the area = 1)')
plt.xlim(0,1)
plt.ylim(0,1.6)
plt.xticks(np.arange(0,  1, step = 0.1)) # setting appropriate labels for x-axis
plt.yticks(np.arange(0,1.6, step = 0.1)) # setting appropriate labels for y-axis
plt.legend(loc='lower center')
plt.grid(c='#CCCCCC')
```
